```python
from transformers import AutoTokenizer, AutoModel, AutoConfig

model_name = "csebuetnlp/banglabert"

tokenizer = AutoTokenizer.from_pretrained(model_name)

config = AutoConfig.from_pretrained("/content/content/training_with_callbacks/checkpoint-9500/config.json")

model = AutoModelForSequenceClassification.from_pretrained("/content/content/training_with_callbacks/checkpoint-9500/pytorch_model.bin",config=config)

model
```

```python
df_test = pd.read_csv("test.csv")
```

```python
df_test
```

```python
pipeline = TextClassificationPipeline(model=model,tokenizer=tokenizer,return_all_scores=True)
output = pipeline(df_test["text"].tolist())
print(output)
```

```python
results = output
```

```python
for res in results:
  print(res)
```

```python
labels = []
for res in results:

  res = [r["score"] for r in res]

  res = np.array(res)
  label = np.argmax(res)
  labels.append(label)
```

```python
df_test['label'] = labels
```

```python
df_test.to_csv("test_predicted.csv")
```

```python
import pickle
from transformers import BertTokenizer, BertForSequenceClassification
from sklearn.preprocessing import LabelEncoder

# Load the pre-trained mBERT-uncased model
model = BertForSequenceClassification.from_pretrained('bert-base-multilingual-uncased')
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-uncased')

new_text = ["রাশিয়ার রাজধানীর নাম মস্কো"]

input_features = tokenizer(new_text, padding="max_length", truncation=True, return_tensors="pt")

# Make predictions
with torch.no_grad():
    predictions = model(**input_features)

predicted_labels = predictions.logits.argmax(dim=1)
label_names = {0: "Fake", 1: "Real"}
print(predicted_labels)
predicted_label_names = [label_names[label] for label in predicted_labels.tolist()]

print(predicted_label_names)
```
