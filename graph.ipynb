```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data with model names and accuracy values
data = {
    'Model': ['U+SVM', 'B+SVM', 'T+SVM', 'U+B+T', 'C3+SVM', 'C4+SVM', 'C5+SVM', 'C3+C4+C5', 'BanglaBERT','SagorBERT', 'M-BERT-c', 'M-BERT-unc', 'XLM-RoBERTa', 'BLOOM'],
    'F1-Score': [0.93, 0.93, 0.90, 0.94, 0.98, 0.97, 0.96, 0.97, 0.94, 0.95, 0.95,0.96, 0.94, 0.96]
}

# Create a DataFrame
df = pd.DataFrame(data)
fig, ax = plt.subplots(figsize=(10, 6))

# Create a bar chart
df.plot(x='Model', y='F1-Score', kind='bar', legend=False, color='cyan', ax=ax)
plt.title('Overall F1 Scores')
plt.xlabel('Model')
plt.ylabel('F1-Score')
plt.ylim(0, 1)  # Set y-axis limits (0 to 1 for accuracy)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data with model names and accuracy values
data = {
    'Model': ['LR', 'SVM', 'BanglaBERT','SagorBERT', 'M-BERT-c', 'M-BERT-unc', 'XLM-RoBERTa', 'BLOOM'],
    'Accuracy': [0.81, 0.83, 0.85, 0.82, 0.80, 0.83, 0.82, 0.87]
}

# Create a DataFrame
df = pd.DataFrame(data)
fig, ax = plt.subplots(figsize=(10, 6))

# Create a bar chart
df.plot(x='Model', y='Accuracy', kind='bar', legend=False, color='#DFAEB4', ax=ax)
plt.title('Model Accuracies')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.ylim(0, 1)  # Set y-axis limits (0 to 1 for accuracy)
plt.show()
```
